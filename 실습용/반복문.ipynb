```python
#문제2) 문제 1번의 구구단에서 곱이 홀수만 나오도록 출력하시오. 예) 2*1, 2*3, 2*5 ...

시작 = 1
끝 = 10
while 시작 < 끝:
    if 시작%2 == 0:
        print(f"출력 : {시작}")
        
        곱 = 1
        while 곱 < 끝:
            if 곱%2 != 0:
                print(f"{시작} * {곱} = {시작 * 곱}")
            곱 = 곱 + 1
        
    시작 = 시작 + 1
```

```python
#반복문에서 사용되는 기능

break : 강제로 종료
continue : 다음으로 점프
```

```python
변수 = 0
while True:
    변수 += 1
    if 변수 == 5:
        break
    else:
        print(변수)
```

```python
변수 = 0
while 변수 < 5:
    변수 += 1
    if 변수%2 == 0:
        continue
    else:
        print(변수)
    
```
